2. Rel-Plot 함수 (relplot) 기초 활용
- 두 변수 간의 관계(relationship)을 파악할 수 있는 차트를 생성해주는 함수
- 수치형 칼럼 간의 관계 파악
- 기본적으로 scatter chart 형태로 표현

```python
#데이터 가져오기
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset('iris')
tips=sns.load_dataset('tips')
fmri=sns.load_dataset('fmri')
titanic=sns.load_dataset('titanic')
flights=sns.load_dataset('flights')
```

```python
#RelPlot 기본 활용
#rel-plot은 두 수치형 칼럼 간의 관계를 파악할 때 사용된다. (관계: relation)
#기본적으로 scatter chart 형태로 표현된다.
print(iris.head())

sns.relplot(data=iris,x='sepal_length',y='petal_length');
```

```python
#set_style: 그림의 전반적인 모양 스타일링
#style:whitegrid #background  color - white, grid 추가
sns.set_style('whitegrid')
sns.relplot(data=iris,x='sepal_length',y='petal_length', hue='species');
#hue는 색깔 구분 가능, 변수를 색칠하여 한 가지 차원을 더 추가
```

```python
sns.set_style('white')
sns.relplot(data=iris,x='sepal_length',y='petal_length', hue='species');
```

```python
sns.set_style('dark')
sns.relplot(data=iris,x='sepal_length',y='petal_length', hue='species');
```

```python
sns.set_style('darkgrid')
sns.relplot(data=iris,x='sepal_length',y='petal_length', hue='species');
```

```python
sns.set_style('ticks')
sns.relplot(data=iris,x='sepal_length',y='petal_length', hue='species');
```

```python
#scatter chart 형태 + 선형 차트, 분산 정보 표시 가능
sns.relplot(data=iris,x='sepal_length',y='petal_length',kind='line');
#kind='scatter'가 default
```

```python
#RelPlot 정리하기
sns.relplot(data=iris,x='sepal_length',y='petal_length',hue='species',kind='line');
```

```python
#RelPlot 정리하기
sns.relplot(data=iris,x='sepal_length',y='petal_length',hue='species');
```

```python
#Tips 데이터에 대하여 total_bill 컬럼과 tip 컬럼에 대한 관계를 차트로 분석하시오
print(tips.head())

sns.relplot(data=tips,x='total_bill',y='tip');
```

```python
#위 분석에서 흡연여부(smoker 컬럼)에 따라 색상을 변경하시오
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker');
```

```python
#라인 차트로 표현하시오
sns.relplot(data=tips,x='total_bill',y='tip',kind='line');
#x축의 동일 값이 존재하지 않을 경우, 분산 영역은 표현되지 않음
```

4. Dist-Plot 함수 (distplot) 기초 활용
- 히스토그램: 수치형 변수 데이터의 분포

```python
#승객들의 탑승 요금 분석 - fare
import pandas as pd
import seaborn as sns

승객데이터=pd.read_csv('titanic.csv')
print(승객데이터.head()) #앞서 csv로 저장하고 다시 불러오니, Unnamed 컬럼이 추가됨.

sns.displot(승객데이터.fare);
```

```python
#그래서 그냥 seaborn 내제 데이터 titanic 사용하겠음
import seaborn as sns
titanic=sns.load_dataset('titanic')
sns.distplot(titanic.fare);
```

```python
#distplot - query() 함수로 필터링
filtered=titanic.query('fare<=150')
sns.distplot(filtered.fare);
```

```python
#승객들의 나이 분석
sns.set_style('whitegrid')
sns.distplot(titanic.age);
```

```python
sns.set_style('whitegrid')
sns.distplot(titanic.age.dropna(), bins=20);
#dropna - na 무시, bins 히스토그램 개수
```

```python
#히스토그램 차트 distribution -> distplot(데이터.컬럼);
sns.distplot(iris.sepal_length); 
```

```python
sns.distplot(iris.sepal_length, bins=20); 
```

6. Bar-Plot 함수 (barplot) 기초 활용
- x축: 하나의 범주형 변수
- y축: 수치형 변수

```python
#성별 생존률 분석
sns.barplot(data=titanic, x='who',y='survived');
```

```python
#좌석 등급별 생존률 분석
sns.barplot(data=titanic, x='class',y='survived');
```

```python
sns.barplot(data=iris, x='species',y='sepal_length');
```

```python
#수평 막대 차트 - orient='h'
#방향 설정 방식 orient -> 'h' 수평, 'v' 수직
#x,y 축의 위치 변경해야 함!!!!!

sns.barplot(data=iris,y='species',x='sepal_length',orient='h');
```

```python
#월별 항공사 승객수
print(flights.head())
sns.barplot(data=flights,x='month',y='passengers');
```

```python
sns.barplot(data=flights,y='month',x='passengers',orient='h');
```

```python
import matplotlib.pyplot as plt
#그래프 사이즈 설정 - plt.figure() 메소드에 figsize 옵션
#figsize(가로길이,세로길이)
plt.figure(figsize=(12,6))
#plt.grid() #grid가 안그려져서
plt.grid(True) #True을 넣어봄
sns.barplot(data=flights,x='month',y='passengers');
```

7. Box-Plot 함수(boxplot) 기초 활용
- 방식1: 1개의 수치형 변수를 입력하여 분석
- 방식2: 1개의 범주형 변수, 1개의 수치형 변수를 입력하여 분석

```python
#1개의 수치형 변수 분석
sns.boxplot(data=iris,x='sepal_length');
```

```python
#식당 지불액 분석
sns.boxplot(data=tips,x='total_bill');
```

```python
sns.boxplot(data=tips,y='total_bill',orient='v');
```

```python
#2개 변수 (x축: 범주, y축: 수치)
print(tips.head())
sns.boxplot(data=tips, x='day',y='total_bill');
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='smoker');
```
