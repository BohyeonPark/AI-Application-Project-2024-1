```python
#우리 동네의 0~100세 이상까지의 인구수를 순서대로 저장해본다.
#3번째 인덱스부터 끝까지 데이터를 반복해서 읽어오면 된다.
import csv
file=open('age.csv',encoding='cp949')
age=csv.reader(file)

for row in age:
  if '동탄5동' in row[0]:
    for i in row[3:]:
      print(i)
```

```python
#앞서 불러온 데이터를 순서대로 저장한다.
#순서대로 저장할 때는 리스트를 사용하는 것이 좋다.
#result라는 빈 리스트를 하나 만들고 각 값을 추가한 후 출력해본다.
import csv
file=open('age.csv',encoding='cp949')
age=csv.reader(file)
result=[] #비어 있는 리스트 생성

for row in age:
  if '동탄5동' in row[0]: #동탄5동이 포함된 행정구역 찾기
    for i in row[3:]: #0세부터 끝까지(100세 이상) 모든 연령에 대해 반복 실행하기
      result.append(i) #인구수를 리스트에 순서대로 추가하기
print(result) #0세부터 100세까지의 인구수 출력하기
```

```python
#출력 결과를 보니 result 리스트에 0~100세 이상까지의 인구수가 잘 저장된 것을 확인 가능하다.
#그런데 각 숫자에 작은따옴표가 있는 것을 보아 문자열로 저장된 것이다.
#데이터를 시각화하려면 각 값을 정수로 변환해야 하므로 수정한다. -> int()
import csv
file=open('age.csv',encoding='cp949')
age=csv.reader(file)
result=[] #비어 있는 리스트 생성

for row in age:
  if '동탄5동' in row[0]: #동탄5동이 포함된 행정구역 찾기
    for i in row[3:]: #0세부터 끝까지(100세 이상) 모든 연령에 대해 반복 실행하기
      result.append(int(i)) #인구수를 리스트에 순서대로 추가하기
print(result) #0세부터 100세까지의 인구수 출력하기
```

```python
#출력 결과를 보니 result 리스트에 0~100세 이상까지의 인구수가 잘 저장된 것을 확인 가능하다.
#그런데 각 숫자에 작은따옴표가 있는 것을 보아 문자열로 저장된 것이다.
#데이터를 시각화하려면 각 값을 정수로 변환해야 하므로 수정한다. -> int()

#데이터 숫자 4자리 이상인 경우 , 콤마 때문에 int 변환 오류 발생하여, 콤마를 공백으로 대체 !!!
import csv
file=open('age.csv',encoding='cp949')
age=csv.reader(file)
result=[] #비어 있는 리스트 생성

for row in age:
  if '동탄5동' in row[0]: #동탄5동이 포함된 행정구역 찾기
    for i in row[3:]: #0세부터 끝까지(100세 이상) 모든 연령에 대해 반복 실행하기
      result.append(int(i.replace(',',''))) #인구수를 리스트에 순서대로 추가하기 #데이터 숫자 4자리 이상인 경우 , 콤마 때문에 int 변환 오류 발생하여, 콤마를 공백으로 대체
print(result) #0세부터 100세까지의 인구수 출력하기
```

2. 인구구조 다양한 형태로 시각화하기
- 원하는 지역의 이름(또는 지역코드)을 입력하면 그 지역의 인구 구조가 출력되게 하기
- 어떤 지역의 인구 구조 그래프인지 알아볼 수 있게 제목을 넣기

```python
#from google.colab import files
#uploaded = files.upload()

import csv
file=open('age.csv',encoding='cp949')
age=csv.reader(file)
result=[]
name=input('인구 구조를 알고 싶은 행정구역이나 행정구역 코드를 입력하세요!!\n')

for row in age:
  if name in row[0]: 
    for i in row[3:]: 
      result.append(int(i))

import matplotlib.pyplot as plt
plt.style.use('ggplot') #격자 무늬로 시각화
plt.rc('font',family='NanumBarunGothic')
plt.title(name + '지역의 인구구조')
plt.plot(result,color='magenta',label='인구수')
plt.xlabel('나이')
plt.ylabel('인구수')
plt.legend()
plt.show()
```

```python
#인구 데이터가 4자리 이상인 경우, ',' 문자열 때문에 int()함수에서 오류가 발생한다.
#i.replace(',','') 수정 필요
#replace(문자열1, 문자열2) i문자열에서 문자열1이 있으면, 문자열2로 대체시키는 함수
#',' 문자열을 '(비어있는 문자열)'로 대체시켜서 지우는 코드

#from google.colab import files
#uploaded = files.upload()

import csv
file=open('age.csv',encoding='cp949')
age=csv.reader(file)
result=[]
name=input('인구 구조를 알고 싶은 행정구역이나 행정구역 코드를 입력하세요!!\n')

for row in age:
  if name in row[0]: 
    for i in row[3:]: 
      result.append(int(i.replace(',','')))

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='NanumBarunGothic')
plt.title(name + '지역의 인구구조')
plt.plot(result,color='magenta',label='인구수')
plt.xlabel('나이')
plt.ylabel('인구수')
plt.legend()
plt.show()
```

3. 인구구조 다양한 형태로 시각화하기
- 성별이 포함된 데이터 다운로드  (행정안전부 홈페이지 - 주민등록 인구 통계, 연령별 남녀구분 인구현황황)
 - 남성 데이터 시작: 인덱스 3~103
 - 여성 데이터 시작: 인덱스 106~206
- 항아리 모양 그래프로 표현하기
- 꺽은선 그래프로 표현하기
- 파이차트로 표현하기기

```python
#항아리 모양 그래프로 표현하기
from google.colab import files
uploaded = files.upload()

import csv
file=open('age_gender.csv',encoding='cp949')
age=csv.reader(file)

m=[]
f=[]

name=input('인구 구조를 알고 싶은 행정구역이나 행정구역 코드를 입력하세요!!\n')

for row in age:
  if name in row[0]: 
    for i in row[3:104]: #남성 데이터 리스트 m에 저장 
      m.append(int(i.replace(',','')))
    for i in row[106:]: #여성 데이터 리스트 f에 저장, row[106:207]
      f.append(int(i.replace(',','')))

print(m)
print(f)

import matplotlib.pyplot as plt
plt.rc('font',family='NanumBarunGothic')
plt.title(name + '지역의 인구구조')
plt.barh(range(101),m) #수평 막대그래프 시각화 barh #수직 막대그래프 시각화 bar
plt.barh(range(101),f) #barh(막대표시할위치, 막대높이) #range()함수를 활용하여 오름차순으로 표현 0~100
plt.show()
```

```python
#항아리 모양 그래프로 표현하기
#그래프가 겹쳐지기 때문에 알아보기가 어렵다 - 두 데이터가 모두 양수로 이어져 있기 때문이다.
#남성 데이터를 왼쪽에, 여성 데이터를 오른쪽에 두려면 남성 데이터를 음수로 바꾸면 된다.

#from google.colab import files
#uploaded = files.upload()

import csv
file=open('age_gender.csv',encoding='cp949')
age=csv.reader(file)

m=[]
f=[]

name=input('인구 구조를 알고 싶은 행정구역이나 행정구역 코드를 입력하세요!!\n')

for row in age:
  if name in row[0]: 
    for i in row[3:104]: #남성 데이터 리스트 m에 저장 
      m.append(-int(i.replace(',',''))) #남성 데이터를 음수로 수정
    for i in row[106:]: #여성 데이터 리스트 f에 저장, row[106:207]
      f.append(int(i.replace(',','')))

print(m)
print(f)

import matplotlib.pyplot as plt
plt.rc('font',family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus']=False #x축 라벨링 마이너스 기호
plt.title(name + '지역의 인구구조')
plt.barh(range(101),m,label='남성') #수평 막대그래프 시각화 barh() #수직 막대그래프 시각화 bar()
plt.barh(range(101),f,label='여성')
plt.legend()
plt.show()
```

```python
#꺽은선 그래프로 표현하기기
#from google.colab import files
#uploaded = files.upload()

import csv
file=open('age_gender.csv',encoding='cp949')
age=csv.reader(file)

m=[]
f=[]

name=input('인구 구조를 알고 싶은 행정구역이나 행정구역 코드를 입력하세요!!\n')

for row in age:
  if name in row[0]: 
    for i in row[3:104]: #남성 데이터 리스트 m에 저장 
      m.append(int(i.replace(',',''))) 
    for i in row[106:]: #여성 데이터 리스트 f에 저장, row[106:207]
      f.append(int(i.replace(',','')))

print(m)
print(f)

import matplotlib.pyplot as plt
plt.rc('font',family='NanumBarunGothic')
plt.title(name + '지역의 인구구조')
plt.plot(range(101),m,label='남성') #꺽은선 그래프 시각화 plot()
plt.plot(range(101),f,label='여성')
plt.legend()
plt.show()
```

```python
#파이차트로 표현하기
#from google.colab import files
#uploaded = files.upload()

import csv
file=open('age_gender.csv',encoding='cp949')
age=csv.reader(file)

size=[] #남여 인구수 저장할 빈 리스트

name=input('인구 구조를 알고 싶은 행정구역이나 행정구역 코드를 입력하세요!!\n')

for row in age:
  if name in row[0]: 
    m=0 #남성 인구수 누적해서 더할 변수 초기화
    f=0 #여성 인구수 누적해서 더할 변수 초기화
    for i in range(101):
      m+=int(row[i+3].replace(',','')) #남성 인구수 누적해서 합 구하기
      f+=int(row[i+106].replace(',','')) #여성 인구수 누적해서 합 구하기

size.append(m)
size.append(f)

print(size)

import matplotlib.pyplot as plt
plt.rc('font',family='NanumBarunGothic')
plt.title(name + '지역의 남여 성별 인구분포')
plt.pie(size, labels=['남','여'], colors=['crimson','green'], autopct='%.1f%%', startangle=90)
#파이 차트 시각화 pie(), 전체 데이터 중 특정 데이터를 보기 쉽게 표현 가능, 입력한 데이터 크기에 따라 원의 비율이 결정됨
#colors 색상 설정, 리스트에 색상 표기
#autopct automation percent 원의 분포 비율 자동으로 표시, %.1f%% 소수점 첫째자리 까지 반올림 숫자 표현 + %기호
#startangle 파이차트의 시작 각도 설정, 90은 반시계 방향으로 90도 이동
plt.legend()
plt.show()
```
