Matplot 라이브러리 개요
- 파이썬으로 데이터를 시각화할 때 가장 많이 사용하는 라이브러리
- 2D 형태의 그래프, 이미지 등을 그릴 때 사용
- 실제 과학 컴퓨팅 연구 분야나 인공지능 연구 분야에서도 많이 활용
- https://matplotlib.org/

Matplot 라이브러리 개요: 그래프나 그림을 그릴 수 있는 플라팅 (plotting) 라이브러리
- 파이썬에서 데이터를 차트나 플롯으로 그려주는 라이브러리 패키지로서 가장 많이 사용되는 데이터 시각화(data visualization) 패키지로 알려져 있다
- 파이썬에서 제공하는 함수들을 사용하면 매트랩(Matlab)처럼 다룰 수 있으며 제공되는 기능이나 문법도 비슷함
- 파이썬에서 Matplot의 기능을 인터렉티브(interactive)하게 사용할 수 있음
- Pyplot 모듈을 사용하여 데이터를 시각적으로 표현
- Matplot과 인터페이스 역할을 하는 것으로 상태 머신(state machine)이라고도 함..!!!!!!!!

axis() 메소드: 그래프의 여백을 할당
- x축과 y축 범위 지정
- [xmin, xmax, ymin, ymax]

```python
x=[1,2,3,3,4]
y=[1,4,9,16,25]

plt.plot(x,y,'bo')
plt.axis([0,6,0,30]) #xmin, xmax, ymin, ymax
plt.show()
```

마커모양 변경
- 마커모양 설정시, 선, 점, 동그라미, 삼각형, 별 모양 다양함
- 마커모양 동시 설정 시 <색상><마커모양><선모양> 순서 입니다.

```python
import matplotlib.pyplot as plt

plt.title("Marker")
plt.plot([10,20,30,40],'r.',label='circle') #r. -> red 동그라미
plt.plot([40,30,20,10],'g^',label='triangle up') #g^ -> green 삼각형
plt.legend()
plt.show()
```

```python
#연습
import matplotlib.pyplot as plt
plt.title("Marker")
plt.plot([0,1,2,3],[10,20,30,40],'r.')
plt.plot([0,1,2,3],[40,30,20,10],'g^')
plt.legend(['circle','triangle up'],loc='center right');
```

Matplotlib 다양한 그래프 함수
- Matplotlib은 위에서 예시한 라인 플롯 이외에 여러 다양한 차트/플롯을 그릴 수 있는데, 각 차트/플롯 마다 다른 함수들을 사용한다.
- 예를 들어, 막대 차트를 그리기 위해서는 plt.bar() 함수를 사용하고, Pie 차트를 그리기 위해서는 plt.pie()를, 히스토그램을 그리기 위해선 plt.hist() 함수를 사용한다.

```python
#막대그래프 plt.bar()
#bar(막대를 표시할 위치, 막대의 높이)
import matplotlib.pyplot as plt
plt.title("bar style")
x=[1,3,5,7,9]
y=[5,7,6,1,4]
plt.bar(x,y) 
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title("bar style")
x=[1,3,5,7,9]
y=[5,7,6,1,4]
plt.barh(x,y) #barh 수평 막대 그래프 horizontal
plt.show()
```

```python
#원형 차트 plt.pie()
import matplotlib.pyplot as plt

plt.title('Subject Score')
data=[90,60,95,100]
label=['kor','math','eng','computer']
plt.pie(data,labels=label) #pie()함수는 labels가 s 복수 형태임!!!
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('Subject Score')
data=[90,60,95,100]
label=['kor','math','eng','computer']
plt.axis('equal') #축의 범위와 축의 스케일을 동일하게 설정하기
plt.pie(data,labels=label)
plt.show()
```

Matplotlib plot의 다양한 서식에 관한 메소드
- legend(): legend 추가
- axis(): 축 교정
- xlabel(), ylabel(): 축의 레이블 설정
- savefig(): 그림 저장
