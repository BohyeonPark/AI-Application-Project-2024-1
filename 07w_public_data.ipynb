```python
import csv 
file = open('rainfall_seoul (1).csv', encoding='cp949') 
rainfall = csv.reader(file) 

next(rainfall) 

max_date='' #가장 비가 많이 온 날짜를 저장할 변수
max_rain=0 #최고 강수량 저장할 변수

for row in rainfall:
  if row[2] == "": #빈 값을 0으로 대체
    row[2]=0

  row[2] = float(row[2]) 

  if max_rain < row[2]: #만약 지금까지 최고 강수량보다 강수량이 많다면
    max_date = row[0] #최고 강수량인 날짜 업데이트
    max_rain = row[2] #최고 강수량 및 업데이트

print("2022년 서울에서 비가 가장 많이 내린 날은 " + max_date+"일이고, 강수량은 " + str(max_rain) + "mm입니다.")

# + 기호는 문자열과 문자열만 연결 가능
# 내가 적은 띄어쓰기로 적용하려면 + 기호, 아니면 , 콤마 사용해도 됨
print("2022년 서울에서 비가 가장 많이 내린 날은 ", max_date, "일이고, 강수량은 ", str(max_rain), "mm입니다.")
```

실습 <서울의 최고 기온 데이터 분석하기>

- 기상자료개방포털 > 기후통계분석 > 기온분석
- 2023년 1월 1일 ~ 2023년 4월 17일
- 지역: 서울

```python
from google.colab import files
uploaded = files.upload()

import csv 
file = open('temp_seoul.csv', encoding='cp949') 
temp = csv.reader(file) #csv 모듈 불러오기

next(temp) 

max_date='' #가장 더웠던 날짜 저장할 변수
max_temp=0 #최고 기온을 저장할 변수

for row in temp:
  if row[4] == "":
    row[4]=-10 #빈 null 값을 최고 기온이 될 수 없는 -10 영하로 설정

  row[4] = float(row[4]) #문자열로 저장된 최고 기온 실수로 변환환

  if max_temp < row[4]: #만약 지금까지 최고 기온보다 기온이 높다면
    max_date = row[0] #최고 기온인 날짜 업데이트
    max_temp = row[4] #최고 기온 및 업데이트

print("2022년 서울에서 최고 기온이 높았던 가장 더웠던 날은" + max_date+"일이고, 기온은" + str(max_temp) + "℃입니다.")
```
