```python
#어린이를 위한 놀이기구 탑승 검사
MAXNUM=4
MAXHEIGHT=130
more=True
cnt=0

while more:
  height=float(input("키는?"))
  if height < MAXHEIGHT:
    cnt+=1
    print('들어가세요.','%d명'%cnt)
  else:
    print("커서 못 들어갑니다.")
  if cnt==MAXNUM:
    more=False
else:
  print("%d명 모두 찼습니다. 다음 번에 이용하세요." %cnt)
```

```python
#다음을 참고로 1에서 99까지 정수인 난수를 2개 생성해 곱하기의 결과를 출력하는 프로그램을 작성하시오.
#반복 while True: 로 계속 난수 발생하여 출력
#표준 입력으로 y 또는 n을 입력 받아 n이면 반복을 종료하고 프로그램을 마침
import random
while True:
  num1=random.randint(1,99)
  num2=random.randint(1,99)
  result=num1*num2
  print('%d X %d = %d' %(num1, num2, result))
  print()
  more=input("계속 y/n?")
  if more in 'nN:':
    break
```

```python
#정수 1개를 표준 입력을 받아 소수인지를 판별하는 프로그램을 작성하시오.
#2 이상의 정수를 입력하여 입력한 정수의 판별
#예) 소수(prime number)인지를 판별한 2이상의 정수 입력>>11
#정수 11는 소수입니다.
n=int(input("소수(prime number)인지를 판별한 2이상의 정수 입력>>"))
cnt=2
for i in range(2,n//2+1):
  if n%i==0:
    cnt+=1
    break
if cnt==2:
  print("정수 %d는 소수입니다." %n)
else:
  print("정수 %d는 소수가 아닙니다." %n)
```
