<텍스트 분석>
1. 자연어처리(natural language processing, NLP)
  - 자연어(natural language)란 우리가 일상생활에서 사용하는 언어를 말합니다.
  - 자연어 처리(natural language processing)란 이러한 자연어의 의미를 분석하여 컴퓨터가 처리할 수 있도록 하는 일을 말한다.
  - 자연어 처리는 음성 인식, 내용 요약, 번역, 사용자의 감성 분석, 텍스트 분류 작업(스팸 메일 분류, 뉴스 기사 카테고리 분류), 질의 응답 시스템, 챗봇과 같은 곳에서 사용되는 분야
  - 자연어 처리는 기계에게 인간의 언어를 이해시킨다는 점에서 인공지능에 있어서 가장 중요한 연구 분야

2. 텍스트 전처리(Text preprocessing)
  - 자연어 처리에 있어서 텍스트 전처리는 매우 중요한 작업입니다. 텍스트 전처리는 용도에 맞게 텍스트를 사전에 처리하는 작업입니다. (요리에 있어서 재료를 제대로 손질하지 않으면, 요리가 제대로 되지 않는 것과 같습니다.)
  - 텍스트에 대해서 제대로 된 전처리를 하지 않으면 뒤에서 배울 자연어 처리 기법들이 제대로 동작하지 않습니다.
  - 토큰화, 정제, 정규화, 어간 추출, 정규식 표현, 인코딩, 데이터 분리, 한국어 전처리 패키지 등

3. 토큰화(Tokenization)
  - 자연어 처리에서 크롤링 등으로 얻어낸 코퍼스(말뭉치) 데이터가 필요에 맞게 전처리되지 않은 상태라면, 해당 데이터를 사용하고자 하는 용도에 맞게 토큰화(tokenization) & 정제(cleaning) & 정규화(normalization)하는 일을 하게 됩니다.
  - 토큰(token)이라 불리는 단위로 나누는 작업을 토큰화(tokenization)라고 부릅니다. 토큰의 단위가 상황에 따라 다르지만, 보통 의미있는 단위로 토큰을 정의합니다.
  - 토큰의 기준을 단어(word)로 하는 경우, 단어 토큰화(word tokenization)라고 합니다. 다만, 여기서 단어(word)는 단어 단위 외에도 단어구, 의미를 갖는 문자열로도 간주되기도 합니다.

4. 단어 토큰화(Word Tokenization)
  - 이 예제에서 토큰화 작업은 굉장히 간단합니다. 구두점을 지운 뒤에 띄어쓰기(whitespace)를 기준으로 잘라냈습니다. 하지만 이 예제는 토큰화의 가장 기초적인 예제
  - 입력: it's beautiful day
  - 출력: ["it","'s","beautiful","day"]
  - 보통 토큰화 작업은 단순히 구두점이나 특수문자를 전부 제거하는 정제(cleaning) 작업을 수행하는 것만으로 해결되지 않습니다. 구두점이나 특수문자를 전부 제거하면 토큰이 의미를 잃어버리는 경우가 발생하기도 합니다.

```python
#형태소 분석기 실습
from konlpy.tag import Okt

txt="오늘학교가는길이아름답습니다."
#인스턴스 생성
#인스턴스: 클래스를 사용하기 위해 메모리 구조를 만든다.
okt=Okt()

#형태소 분석
okt.pos(txt)
```

```python
from konlpy.tag import Okt

txt2="아버지가방에들어가신다."
okt=Okt()
okt.pos(txt2)
```

```python
okt.pos("신나용ㅋㅋㅋ")
```

```python
#stem=True: 원형 글자로 바꿔준다. (기본값 False)
#오타 및 잘못 적은 글자 변환 가능함
okt.pos("신나용ㅋㅋㅋ", norm=True, stem=True)
```

```python
#morphs() 텍스트를 형태소 단위로 나눈다.
okt.morphs(txt)
```

```python
#nouns() 텍스트에서 명사만 추출
okt.nouns(txt)
```

```python
#KoNLPy Kkma 임포트
from konlpy.tag import Kkma

#텍스트에서 문장을 분석
kkma=Kkma()
kkma.sentences(txt)

#txt 문장에서 띄어쓰기 등의 문장을 분석해줌
```

```python
import konlpy
text="요즘 인공지능, 머신러닝, 티처블머신, 데이터 분석이 너무 잼있어요 ㅋ, 하지만 어렵기도 합니닿ㅎㅎㅎ.~"

okt=konlpy.tag.Okt()
print(okt.morphs(text))
print(okt.morphs(text,stem=True))
#stem=True: 원형글자로 바꿔준다. 기본값 False
```

```python
#품사 알아보기
print(okt.pos(text, stem=True))
```

```python
#특정품사 찾기
print(okt.nouns(text))
```

```python
#명사, 동사, 형용사 찾기
okt=konlpy.tag.Okt()

for word in okt.pos(text):
  if word[1] in ['Noun','Verb','Adjective']: #명사, 동사, 형용사
    print(word[0])
```

```python
from konlpy.tag import Okt
okt=Okt()

for t in okt.pos(text):
  if t[1] in ['Noun','Verb','Adjective']:
    print(t[0])
```
